## Creating Vectordb

```python
import re
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

def process_verse(verse):
    # Extract chapter and verse number using regex
    match = re.match(r"(\d+)\|(\d+)\|(.+)", verse)
    if match:
        chapter_no = match.group(1)
        verse_no = match.group(2)
        verse_text = match.group(3).strip()
        return chapter_no, verse_no, verse_text
    else:
        raise ValueError(f"Verse format is incorrect: {verse}")

# Convert each verse into a Document object with metadata
source_docs = []
for verse in quran_verses:
    chapter_no, verse_no, verse_text = process_verse(verse)
    source_docs.append(
        Document(
            page_content=verse_text,
            metadata={"chapter": chapter_no, "verse": verse_no}
        )
    )

# Initialize the embedding model
embedding_model = HuggingFaceEmbeddings(model_name="thenlper/gte-small")

# Store the documents in a FAISS vector database
vectordb = FAISS.from_documents(documents=source_docs, embedding=embedding_model)
```

## Retrieving relevant verses

```python
import numpy as np

def query_faiss_index(query, vectordb, embedding_model, num_results):
    # Ensure the query is a single string
    if isinstance(query, str):
        query = [query]  # Convert to list for embedding
    elif isinstance(query, list) and all(isinstance(q, str) for q in query):
        query = query
    else:
        raise ValueError("Query must be a string or a list of strings.")

    # Generate embeddings for the query
    query_embedding = embedding_model.embed_query(query[0])

    # Perform the search in the FAISS index
    results = vectordb.similarity_search(query[0], k=num_results)

    # Process the results to include metadata
    formatted_results = []
    for result in results:
        # Extract the verse text and metadata
        verse_text = result.page_content
        chapter_no = result.metadata.get('chapter')
        verse_no = result.metadata.get('verse')

        # Format the result
        formatted_verse = f"Quran ({chapter_no}:{verse_no}) {verse_text}"
        formatted_results.append(formatted_verse)

    return formatted_results
```

## Preparing the context

```python
def prepare_context(documents):
    return "\n".join([doc for doc in documents])
```

## Generating response

```python
from transformers import pipeline

def generate_response(query, context):
    # Combine the context and the query
    input_text = f"Context:\n{context}\n\nFrom the context given above, answer the following question concisely:\n{query}\nAnswer:"

    # Initialize the pipeline
    pipe = pipeline("text-generation", model="openai-community/gpt2-xl",device=0, max_length=400, temperature=0.8)

    # Generate the output
    output = pipe(input_text, max_new_tokens=100, truncation=True)

    # Extract just the answer part from the generated text
    response = output[0]['generated_text'].split('\nAnswer:')[1].strip() if '\nAnswer:' in output[0]['generated_text'] else output[0]['generated_text'].strip()

    return response
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# Initialize the embedding model
embedding_model = HuggingFaceEmbeddings(model_name="thenlper/gte-small")

def rag_pipeline(query, vectordb,embedding_model, num_results=3):
    # Step 1: Retrieve relevant documents
    retrieved_docs = query_faiss_index(query, vectordb, embedding_model, num_results)

    # Step 2: Prepare the context
    context = prepare_context(retrieved_docs)

    # Step 3: Generate the response
    response = generate_response(query, context)

    return response
```

```python
# Example usage
query = "What is the mother of christ?"
response = rag_pipeline(query, vectordb, embedding_model)
print(response)
```

```python
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel, Field
from typing import List
import nest_asyncio
from pyngrok import ngrok
import uvicorn
import threading

# Assuming the rag_pipeline function and required models are already defined
# from your_rag_module import rag_pipeline

app = FastAPI()

class QueryRequest(BaseModel):
    query: str = Field(..., title="Query", max_length=500, example="What does the Quran say about patience?")

@app.post("/ask")
def ask_question(query_request: QueryRequest):
    query = query_request.query
    try:
        # Call the RAG pipeline with the provided query
        response = rag_pipeline(query, vectordb, embedding_model)
        return {"response": response}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

# Start the FastAPI server
def start_server():
    uvicorn.run(app, host="127.0.0.1", port=8001)

# Run the FastAPI server in a separate thread
def run_fastapi_in_thread():
    thread = threading.Thread(target=start_server)
    thread.start()

# Required to run the server in Colab
nest_asyncio.apply()
ngrok_tunnel = ngrok.connect(8001)
print('Public URL:', ngrok_tunnel.public_url)

# Start the FastAPI server in a background thread
run_fastapi_in_thread()
```

```python
import gradio as gr
import requests

# Assuming the rag_pipeline function and required models are already defined
# from your_rag_module import rag_pipeline

def gradio_interface(query):
    try:
        response = requests.post("http://127.0.0.1:8001/ask", json={"query": query}) #rag_pipeline(query, vectordb, embedding_model)
        if response.status_code == 200:
            return response.json().get('response', 'No response field in JSON')
        else:
            return f"Error: {response.status_code}, {response.text}"
    except Exception as e:
        return str(e)

# Custom CSS for styling
custom_css = """
body {background-color: #f5f5f5; font-family: 'Arial', sans-serif;}
h1 {color: #2c3e50;}
textarea, input {font-family: 'Arial', sans-serif; font-size: 16px;}
"""

# Gradio Interface
iface = gr.Interface(
    fn=gradio_interface,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Enter your question here...",
        label="Ask a Question"
    ),
    outputs=gr.Textbox(label="Answer"),
    title="Ask the Quran",
    description=(
        "This application allows you to ask questions related to the Quran. "
        "Simply enter your question, and the model will provide an answer based on the Quran's teachings."
    ),
    theme="default",
    css=custom_css,
)

# Launch the interface
iface.launch(share=True)
```
